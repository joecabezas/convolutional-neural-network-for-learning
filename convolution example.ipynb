At this point we have the feature, now lets convolute it on the image

Lets define some hyper variables

```python
stride = 1
```

Now the convolution operation

For the sake of simplicity, lets flatten the output to a single channel by doing a SUM as represented here, but in our case we dont have a filter for each channel, is a single 3d filter (MxNx3)


![](http://machinelearninguru.com/_images/topics/computer_vision/basics/convolutional_layer_1/rgb.gif)

```python
#define the convolution function
def convolute_images(images, a_filter, stride=1):
    #calculate the space for the output of the convolution
    image_count = images.shape[0]
    out_cols = int(np.ceil((images.shape[1] - a_filter.shape[0] + 1)/stride))
    out_rows = int(np.ceil((images.shape[2] - a_filter.shape[1] + 1)/stride))
    channels = images.shape[3]

    #reserve space for output
    output = np.zeros((image_count, out_cols, out_rows, channels))
    
    assert len(images.shape) == 4, "As input is expected an array of images, which are (MxNx3)"
    assert len(a_filter.shape) == 3, "As filter, is expected a filter for an image in form (MxNx3)"

    for i in range(out_cols):
        for j in range(out_rows):
            col_start = i * stride
            col_end = col_start + a_filter.shape[0]
            row_start = j * stride
            row_end = row_start + a_filter.shape[1]

            w_x = images[:, col_start:col_end, row_start:row_end, :] * a_filter
            output[:, i, j, :] = np.sum(
                w_x,
                axis=(1,2)
            )
    #at this point we have the 3-dimensional output, let's flatten it
    return np.sum(output, axis=(3))
```

```python
#define a single channel plot helper
def plot_single_channel_image(image):
    assert len(image.shape) == 2, "Only 2d images"

    fig, ax = plt.subplots()
    cax = ax.imshow(image, interpolation=None, cmap='inferno')
    cbar = fig.colorbar(cax, orientation='vertical')
    plt.show()
```

```python
#now lets do the convolution
output = convolute_images(images, filter_red_spot, stride=1)
```
